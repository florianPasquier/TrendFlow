```python
import pandas as pd
from pandas import json_normalize
import json

with open('google_trends.json') as f:
    data = json.load(f)
    
df = json_normalize(data['trending_searches'])
df['categories'] = df['categories'].apply(lambda x: [d['name'] for d in x])
df.drop(columns=['serpapi_google_trends_link'], inplace=True)
df.to_csv('google_trends.csv', index=False,encoding='utf-8')
df
```

```python
with open("x_trends.json", "r", encoding="utf-8") as f:
    data = json.load(f)
# dfj = pd.read_json('x_trends.json')
# df = json_normalize(data,max_level=0)
# dfj.T
# df.to_csv('x_trends.csv', index=False,encoding='utf-8')
rows = []
for currentness, trends in data.items():
    for trend in trends["trend"]:
        rows.append(
            {"currentness": currentness, "name": trend["name"], "count": trend["count"]}
        )
df = pd.DataFrame(rows)
df.to_csv("x_trends.csv", index=False, encoding="utf-8")
df
```
