```python
import pandas as pd
from pandas import json_normalize
import json

with open("google_trends.json") as f:
    data = json.load(f)

df = json_normalize(data["trending_searches"])
# Flatten the 'categories' column
df["categories"] = df["categories"].apply(lambda x: [d["name"] for d in x])
# Drop the 'serpapi_google_trends_link' column
df.drop(columns=["serpapi_google_trends_link"], inplace=True)
# Add columns 'source' and 'currentness'
df["source"] = "google_trends"
df["currentness"] = pd.to_datetime(df["start_timestamp"], unit="s").dt.strftime(
    "%Y-%m-%d-%H:%M:%S"
)

# Rename column 'query' to 'name'
df = df.rename(
    columns={
        "query": "name",
        "search_volume": "count",
        "trend_breakdown": "associated_entities",
    }
)

# Drop columns: 'increase_percentage', 'end_timestamp'
df = df.drop(
    columns=["increase_percentage", "end_timestamp", "start_timestamp", "active"]
)
df.to_csv("google_trends.csv", index=False, encoding="utf-8", sep=";")
```

```python
import json
import pandas as pd
from datetime import datetime, timedelta

with open("x_trends.json", "r", encoding="utf-8") as f:
    data = json.load(f)
# Flatten the JSON data
rows = []
# Define the format for the timestamp
frm = "%Y-%m-%d-%H:%M:%S"
# Iterate over the data and create a row for each trend
for currentness, trends in data.items():
    for trend in trends["trend"]:
        rows.append(
            {
                "currentness": (
                    datetime.now() - timedelta(hours=int(currentness))
                ).strftime(frm),
                "name": trend["name"],
                "count": trend["count"],
            }
        )
df = pd.DataFrame(rows)
df["source"] = "x_trends"
df["categories"] = ""
df["associated_entities"] = ""
df.to_csv("x_trends.csv", index=False, encoding="utf-8")
df
```
